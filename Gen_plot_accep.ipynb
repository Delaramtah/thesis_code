```python
## correctuin in reco
import awkward as ak

def produce_pairs(electron_reco):
    Electron_charge = ak.combinations(electron_reco["Electron_charge"],2)
    lefts, rights = ak.unzip(Electron_charge)
    mask_Electron = lefts*rights == -1 # they must have opposite charge
    
    # select paris of opposite charged electons
    p_pt = ak.flatten(ak.combinations(electron_reco["Electron_pt"],2)[mask_Electron])
    p_eta = ak.flatten(ak.combinations(electron_reco["Electron_eta"],2)[mask_Electron])
    p_phi = ak.flatten(ak.combinations(electron_reco["Electron_phi"],2)[mask_Electron])
    p_energy = ak.flatten(ak.combinations(electron_reco["Electron_ecalEnergy"],2)[mask_Electron])
        
    # 2.) calculate mass of each pair
    # split the pairs 
    l_pt, r_pt = ak.unzip(p_pt)
    l_eta, r_eta = ak.unzip(p_eta)
    l_phi, r_phi = ak.unzip(p_phi)
    l_energy, r_energy = ak.unzip(p_energy)

    return l_pt,l_eta,l_phi,l_energy,r_pt,r_eta,r_phi,r_energy
```

```python
def produce_pairs(electrons):
    Electron_charge = ak.combinations(electrons["Electron_charge"],2)
    lefts, rights = ak.unzip(Electron_charge)
    mask_Electron = lefts*rights == -1 # they must have opposite charge
    elMass = 0
    # select paris of opposite charged electons
    p_pt = ak.combinations(electrons["Electron_pt"],2)[mask_Electron]
    p_eta = ak.combinations(electrons["Electron_eta"],2)[mask_Electron]
    p_phi = ak.combinations(electrons["Electron_phi"],2)[mask_Electron]
        
    # 2.) calculate mass of each pair
    # split the pairs 
    l_pt, r_pt = ak.unzip(p_pt)
    l_eta, r_eta = ak.unzip(p_eta)
    l_phi, r_phi = ak.unzip(p_phi)

    el1 = vector.obj(pt=l_pt, phi=l_phi, eta=l_eta, mass=l_pt*0+elMass)
    el2 = vector.obj(pt=r_pt, phi=r_phi, eta=r_eta, mass=r_pt*0+elMass)

    return el1, el2

```
